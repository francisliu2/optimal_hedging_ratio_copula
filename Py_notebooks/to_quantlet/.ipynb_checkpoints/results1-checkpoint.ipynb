```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
import sys, os
sys.path.append('../../src/')
from toolbox import *
import datetime

result_names = ['BBT_Tiingo',                #0
                'BBT_future_Tiingo_ada', 
                'BBT_future_Tiingo_eth',     #2
                'BBT_future_Tiingo_ltc',
                'BBT_future_Tiingo_xrp',     #4
                'BBT_future_BITX',
                'BBT_future_BITW20',         #6
                'BBT_future_BITW70', 
                'BBT_future_BITW100',        #8
                'BBT_future_CRIX']

risk_measures = ['Variance',
                 'ES q=0.05', 'ES q=0.01',
                 'VaR q=0.05', 'VaR q=0.01',
                 'ERM k=10', 'CH']

order = ['BBT_Tiingo',    
         'BBT_future_CRIX', 'BBT_future_BITX', 'BBT_future_BITW100', 
         'BBT_future_BITW20',  'BBT_future_BITW70',
         'BBT_future_Tiingo_ada', 'BBT_future_Tiingo_eth',  'BBT_future_Tiingo_ltc','BBT_future_Tiingo_xrp']

coin_names = ['btc', 'CRIX', 'BITX', 'BITW100', 'BITW20', 'BITW70', 'ada', 'eth', 'ltc', 'xrp']

new_coin_names = ['BTC', 'CRIX', 'BITX', 'BITW100', 'BITW20', 'BITW70', 'ADA', 'ETH', 'LTC', 'XRP']
```

```python
# BTC
n = 0
rn = order[n]
rm = 'ES q=0.05'
path = '../../results/%s/MM/rh_%s_copula_selection.csv'\
                %(rn, rm)

df = pd.read_csv(path)
df.Date = pd.to_datetime(df.Date)

df = df.sort_values('Date')
df.rsc = np.cumsum(df.rs)
df.rfc = np.cumsum(df.rf)
df.rhc = np.cumsum(df.rh)

_max_rs = np.max([np.max(df.rsc), np.max(df.rfc)])
_min_rs = np.min([np.min(df.rsc), np.min(df.rfc)])
_y_rs = np.max([np.abs(_max_rs), np.abs(_min_rs)])*1.05

_max_rh = np.max(df.rh)
_min_rh = np.min(df.rh)
_y_rh = np.max([np.abs(_max_rh), np.abs(_min_rh)])*1.5

fig, ax = plt.subplots(figsize=(16,9),nrows=4, sharex=True)
ax[0].plot(df.Date, df.rsc)
ax[1].plot(df.Date, df.rfc)


rs_5min = df.sort_values('rs')[:5]
# ax[0].scatter(list(rs_5min.Date), list(rs_5min.rs), marker='o', color='red')
for k in range(len(rs_5min)):
    ax[0].axvline(x=rs_5min.Date.iloc[k], color='black', linestyle='-', linewidth=0.5)
    ax[2].axvline(x=rs_5min.Date.iloc[k], color='black', linestyle='-', linewidth=0.5)

h_max = np.max(df.h)
h_min = np.min(df.h)

for rm in risk_measures[:-1]:
    path = '../../results/%s/MM/rh_%s_copula_selection.csv'\
                    %(rn, rm)
    
    df = pd.read_csv(path)
    df.Date = pd.to_datetime(df.Date)

    h_max = np.max([h_max, np.max(df.h)])
    h_min = np.min([h_min, np.min(df.h)])
    
    ax[2].plot(df.Date, df.rh, label=rn)
    ax[3].plot(df.Date, df.h)

ax[3].set_ylim((h_min*0.75, h_max*1.25))

ax[0].set_ylabel(r'$r_s$')
ax[1].set_ylabel(r'$r_f$')
ax[2].set_ylabel(r'$r_h$')
ax[3].set_ylabel('OHR')

for i in range(2):
    ax[i].set_ylim((-_y_rs,_y_rs))
    
ax[2].set_ylim((-_y_rh,_y_rh))

start_day = df.Date.iloc[-1]-datetime.timedelta(days=5)
end_day = df.Date.iloc[0]+datetime.timedelta(days=5)

ax[0].set_xlim((start_day, end_day))
    
plt.subplots_adjust(left=None,
                bottom=None, 
                right=None, 
                top=None,
                wspace=None,
                hspace=0.05)

ax[0].set_title(new_coin_names[n])
```

```python
# BTC
n = 0
rn = order[n]
rm = 'ES q=0.05'
train_path = '../../processed_data/%s/train/'\
                %(rn)

test_path = '../../processed_data/%s/test/'\
                %(rn)

# df = pd.read_csv(path)
# df.Date = pd.to_datetime(df.Date)
train_files = [x for x in os.listdir(train_path) if x.endswith('.csv')]

test_files = [x for x in os.listdir(test_path) if x.endswith('.csv')]
```

```python
Full_df = pd.DataFrame()

for i in range(len(train_files)):
    file_path = train_path+train_files[i]
    df        = pd.read_csv(file_path)
    df.Date   = pd.to_datetime(df.Date)
    df.loc[:,'file_tpye'] = 'train'
    Full_df   = Full_df.append(df)
    
for i in range(len(test_files)):
    file_path = test_path+test_files[i]
    df        = pd.read_csv(file_path)
    df.Date   = pd.to_datetime(df.Date)
    df.loc[:,'file_tpye'] = 'test'
    Full_df   = Full_df.append(df)

Full_df = Full_df.sort_values('Date', ascending=True)
```

```python
Full_df = Full_df.drop_duplicates('Date', keep='last')
```
