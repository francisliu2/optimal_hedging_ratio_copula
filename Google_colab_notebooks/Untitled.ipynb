```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import seaborn as sns
import scipy
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
r1 = stats.norm(0,1).rvs(size=350)
r2 = stats.norm(2,0.05).rvs(size=350)
sns.distplot(r1)
sns.distplot(r2)
```

```python
_min = np.min([np.min(r1), np.min(r2)])
_max = np.max([np.max(r1), np.max(r2)])

x = np.linspace(_min, _max, 100)

ecdf1 = ECDF(r1)(x)
ecdf2 = ECDF(r2)(x)

plt.plot(x, ecdf1)
plt.plot(x, ecdf2)
```

```python
S2_1 = np.cumsum(ecdf1)
S2_2 = np.cumsum(ecdf2)
plt.plot(x, S2_1)
plt.plot(x, S2_2)
print(np.mean(S2_1>S2_2))
```

```python
def I(z, X, j):
    # Whole sample
    N = len(X)
    return 1/N*np.sum((X <= z)*(z-X)**(j-1))/np.math.factorial(j-1)

def run(r1, r2, j):
    r1_resampled = []
    r2_resampled = []

    bootstrap_size = 500

    for _ in range(bootstrap_size):
        ids = np.random.choice(range(0,350), size=500, replace=True)
        r1_resampled.append(r1[ids])
        r2_resampled.append(r2[ids])

    S_bar_arr = []
    for i in range(len(r1_resampled)):
        fn = lambda z: I(z, r1_resampled[i], j) - I(z, r2_resampled[i], j)
        fn_opt = lambda z: -(fn(z))**2
        S_bar = fn(scipy.optimize.fmin(fn_opt, 0))
        S_bar_arr.append(S_bar)
    S_bar = np.array(S_bar_arr).reshape((-1))
    
    fn = lambda z: I(z, r1, j) - I(z, r2, j)
    fn_opt = lambda z: -(fn(z))**2
    S_hat = fn(scipy.optimize.fmin(fn_opt, 0))
    
    return S_bar, S_hat
```

```python
S_bar_1, S_hat_1 = run(r1, r2, 2)
S_bar_2, S_hat_2 = run(r2, r1, 2)
```

```python
sns.distplot(S_bar_1)
print(S_hat_1)
```

```python
np.mean(S_bar_1 > S_hat_1)
```
