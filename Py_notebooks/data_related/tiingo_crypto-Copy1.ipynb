```python
dates = pd.date_range(start="2017-12-15", end="2021-06-01", freq="3D")
dates_end = pd.date_range(start="2017-12-18", end="2021-06-04", freq="3D")
```

```python
s = str(dates[0]).split(' ')[0]
e = str(dates_end[0]).split(' ')[0]

print(s,e)
r = requests.get("https://api.tiingo.com/tiingo/crypto/prices?tickers=btcusd&" \
        "startDate=" + s + "&endDate=" + e + "&resampleFreq=1hour", headers=headers)
```

```python
r.json()
```

```python
headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Token ' + token
}
filename="tiingo_btc.csv"
header=True
if os.path.isfile(filename):
    header=False

for d1, d2 in zip(dates, dates_end):
    print(str(d1).split(' ')[0] + ", " + str(d2).split(' ')[0]);
    requestResponse \
        = requests.get("https://api.tiingo.com/tiingo/crypto/prices?tickers=btcusd&" \
        "startDate=" + str(d1).split(' ')[0] + "&endDate=" + str(d2).split(' ')[0] + "&resampleFreq=1hour", headers=headers)
    rj = requestResponse.json()
    df=pd.DataFrame(rj[0]['priceData'])
    df=df[['date', 'open', 'high', 'low', 'close', 'volume', 'tradesDone', 'volumeNotional']]
    df.to_csv(filename, mode="a", header=header)
    header=False
    time.sleep(2)          
```
