## Simulation Study
Motivation: To see how does the parameter estimation looks like if there are structural break points. 

We simulate a time series of bivariate data with structural break points and estimate copula in a rolling window basis. Then, we plot the time series of the estimate against the true parameters.

We simulate 1500 data points in total with two break points at 500th and 1000th. The first 500 data are from a Gaussian copula with $\rho=0.5$, the next with $\rho=0.7$, and the last 500 are with $\rho=0.9$.

For estimation, we adapt a window size of 100 with step size of 1. We label the estimation by the index of the last data point used in calibration. 

```python
sample_rho_arr = [0.5, 0.9, 0.1]
sample_size_per_rho = 500
samples = [Gaussian({'rho':rho}, stats.norm, stats.norm).sample(sample_size_per_rho) for rho in sample_rho_arr]
samples = np.array(samples).reshape((-1,2))
window_size = 100
step_size = 1

rs = samples[:,0]
rf = samples[:,1]
```

```python

rho_arr = []
likelihood_arr = []

C1 = Gaussian(dict(rho=0.8), Law_RS=stats.norm, Law_RF=stats.norm)

for i in range(len(samples)-window_size-step_size):
    _rs = samples[(i*step_size):((i+1)*step_size)+window_size,0]
    _rf = samples[(i*step_size):((i+1)*step_size)+window_size,1]
    
    _u = ECDF(_rs)(_rs)
    _v = ECDF(_rf)(_rf)
    
    C1.canonical_calibrate(_u,_v)
    rho_arr.append(C1.rho)
    likelihood_arr.append(C1.l_fn(C1.rho, _u,_v))
    
true_rho_arr = np.array([[rho]*sample_size_per_rho for rho in sample_rho_arr]).reshape((-1))
```

## Difference in Training and Testing Kendall's $\tau$ 
We compute the Kendall's $\tau$ of training sets and their corresponding testing sets. We can see that the $\tau$ of training sets are differ from that of its testing sets. 

```python
data_name = "future_brr_v4"
data_path = "../processed_data/"+data_name+"/"
ls = os.listdir(data_path+'train/')
ls = [l for l in ls if l.endswith('.csv')]
    
spot_name = "log return brr"
future_name = "log return future"

tau_arr_train = []
tau_arr_test = []

p_value_train = []
p_value_test = []

for i, file in enumerate(ls):
    train = pd.read_csv(data_path+'train/'+file)
    test = pd.read_csv(data_path+'test/'+file)
    
    spot_train   = train.loc[:,spot_name]
    future_train = train.loc[:,future_name]
        
    spot_test   = test.loc[:,spot_name]
    future_test = test.loc[:,future_name]
    
    tau_arr_train.append(stats.kendalltau(spot_train, future_train)[0])
    tau_arr_test.append(stats.kendalltau(spot_test, future_test)[0])
    p_value_train.append(stats.kendalltau(spot_train, future_train)[1])
    p_value_test.append(stats.kendalltau(spot_test, future_test)[1])
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ls, tau_arr_train, label='Train')
plt.scatter(ls, tau_arr_test, label='Test')
plt.legend()
plt.title("Kendall's Tau of train set and test set")
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ls, p_value_train, label='Train')
plt.scatter(ls, p_value_test, label='Test')
plt.legend()
plt.title("P value of Kendall's Tau estimation of train set and test set")
```

```python
mu1 = 0
mu2 = 0
var1 = 1
var2 = 1
cov = 0

def l_fn(theta, data, weights):
    mu1, mu2, var1, var2, cov = theta
    Sigma = np.array([[var1, cov], [cov, var2]])
#     print(Sigma)
    if (var1 <=0) or (var2 <=0):
        return -5000
    if np.linalg.det(Sigma)<=0.1:
        return -5000

    MVN = stats.multivariate_normal([mu1,mu2],Sigma)
    return np.mean(weights*np.log(MVN.pdf(data)))
```

```python
s = 120
X = np.array(data.iloc[:s*2,[4,5]])
# X = samples

X_A = X[:s,:]
X_B = X[s:,:]

def critical_values(X, s):
    X_A = X[:s,:]
    X_B = X[s:,:]
    noise = 1
    fn = lambda theta: -l_fn(theta, X_A, noise)
    theta_A = scipy.optimize.fmin(fn,
                        x0=(0,0,1,1,0), 
                        maxfun=50000,
                        maxiter=50000,disp=False)
    likelihood_A = l_fn(theta_A, X_A, noise)

    fn = lambda theta: -l_fn(theta, X_B, noise)
    theta_B = scipy.optimize.fmin(fn,
                        x0=(0,0,1,1,0), 
                        maxfun=50000,
                        maxiter=50000,disp=False)
    likelihood_B = l_fn(theta_B, X_B, noise)

    fn = lambda theta: -l_fn(theta, X, noise)
    theta = scipy.optimize.fmin(fn,
                        x0=(0,0,1,1,0), 
                        maxfun=50000,
                        maxiter=50000,disp=False)
    likelihood = l_fn(theta, X, noise)
    T = likelihood_A + likelihood_B - likelihood # test statistics

    x0 = np.array([0,0,100,100,0])

    T_circ_arr = [] # Placeholder for multiplier bootstrap test statistics

    for _ in tqdm(range(400)):
        noise = stats.expon(0).rvs(len(X))

        fn = lambda theta: -l_fn(theta, X_A, noise[:s])
        theta_A_circ = scipy.optimize.fmin(fn,
                            x0=x0, 
                            maxfun=50000,
                            maxiter=50000,disp=False)
        likelihood_A_circ = l_fn(theta_A, X_A, noise[:s])

        fn = lambda theta: -l_fn(theta, X_B, noise[s:])
        theta_B_circ = scipy.optimize.fmin(fn,
                            x0=x0, 
                            maxfun=50000,
                            maxiter=50000,disp=False)
        likelihood_B_circ = l_fn(theta_B, X_B, noise[s:])

        fn = lambda theta: -l_fn(theta+theta_B-theta_A, X, noise)
        theta_circ = scipy.optimize.fmin(fn,
                            x0=x0, 
                            maxfun=50000,
                            maxiter=50000,disp=False)
        likelihood_circ = l_fn(theta_circ+theta_B-theta_A, X, noise)

        T_circ = likelihood_A_circ + likelihood_B_circ - likelihood_circ
        T_circ_arr.append(T_circ)
        
    return T, T_circ_arr
```
